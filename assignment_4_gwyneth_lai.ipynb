```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.booking.com/pets/city/hk/hong-kong.en-gb.html"
result = requests.get(url)
soup = BeautifulSoup(result.content,'html.parser')
print(result)
```

```python
header = ['Pet-Friendly Hotels','Type','Rating']
data = []
for item in soup.find_all("div", attrs={"class":"bui-card__content"}):
    Hotel = item.find ("h3",attrs={"class":"bui-card__title"})
    Type = item.find ("span",attrs={"class":"bui-badge bui-badge--outline"})
    Rating = item.find ("span",attrs={"class":"review-score-badge"})
    print(Hotel)
    print(Type)
    print(Rating)
    data.append((Hotel,Type,Rating))

df = pd.DataFrame(data,
   columns = header) 
df.to_csv('booking.com_hotel.csv', sep='\t', encoding='utf-8')
df
```
